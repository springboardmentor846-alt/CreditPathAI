```python
# Reading the data files stored in data/raw/: 
import pandas as pd
import os

RAW_PATH = "C://Users//Rachit//OneDrive//Documents//CreditPathAI//data//raw"
files = os.listdir(RAW_PATH)

files
```

```python
import os 

print("Notebook Location:",os.getcwd())
print("Files in data/raw/:",os.listdir("C://Users//Rachit//OneDrive//Documents//CreditPathAI//data//raw"))
```

```python
datasets = {}

for f in files:
    if f.endswith(".csv"):
        file_path = os.path.join(RAW_PATH, f)
        if os.path.exists(file_path):
            try:
                df = pd.read_csv(file_path, nrows=5, encoding="latin1")
            except UnicodeDecodeError:
                try:
                    df = pd.read_csv(file_path, nrows=5, encoding="ISO-8859-1")
                except UnicodeDecodeError:
                    df = pd.read_csv(file_path, nrows=5, encoding="cp1252")
            
            datasets[f] = df
        else:
            print("File not found:", file_path)

datasets.keys()
```
